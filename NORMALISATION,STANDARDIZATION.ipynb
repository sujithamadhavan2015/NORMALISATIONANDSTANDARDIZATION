# NORMALIZATION
# Data has uneven gap between minimum to maximum value we use normalization.


```python
from sklearn import preprocessing
```

```python
df_norm = preprocessing.normalize(df)
```

```python
df_norm
```

```python
plt.xlabel('df_norm')
plt.plot(df_norm,"*")
```

# L1 (NORMALIZATION)
# We compressing all data between (-1.....0.....1),then the data will beeasy to handle.

```python
from sklearn.preprocessing import Normalizer
l1_normalizer = Normalizer(norm = 'l1').fit(df)             # (norm = 'l1') (method = l1)
```

```python
l1_normalized = l1_normalizer.transform(df)
```

```python
l1_normalized
```

```python
plt.plot(l1_normalized ,"*")
```
